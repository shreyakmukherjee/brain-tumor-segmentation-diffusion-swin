```python
import os
import time
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from timm.models.swin_transformer import SwinTransformer
from torch.cuda.amp import GradScaler, autocast
import cv2
from glob import glob
from sklearn.model_selection import train_test_split
from torchvision.models import mobilenet_v3_small
import pandas as pd
from tqdm import tqdm
import albumentations as A
import matplotlib.patches as mpatches
from sklearn.metrics import confusion_matrix, roc_curve, auc, precision_recall_curve, average_precision_score
import seaborn as sns
from sklearn.metrics import classification_report

# Configuration (unchanged)
class Config:
    img_size = 256
    batch_size = 16
    num_epochs = 50
    lr = 1e-4
    timesteps = 100
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    save_path = "brain_tumor_model.pth"
    data_dir = '/kaggle/input/lgg-mri-segmentation/lgg-mri-segmentation/kaggle_3m'
    csv_path = '/kaggle/input/lgg-mri-segmentation/lgg-mri-segmentation/kaggle_3m/data.csv'

# Data Loading and Preprocessing (unchanged)
def get_file_row(path):
    path_no_ext, ext = os.path.splitext(path)
    filename = os.path.basename(path)
    patient_id = '_'.join(filename.split('_')[:3])
    return [patient_id, path, f'{path_no_ext}_mask{ext}']

def load_data():
    df = pd.read_csv(Config.csv_path)
    file_paths = glob(f'{Config.data_dir}/*/*[0-9].tif')
    filenames_df = pd.DataFrame((get_file_row(f) for f in file_paths), 
                              columns=['Patient', 'image_filename', 'mask_filename'])
    df = pd.merge(df, filenames_df, on="Patient")
    train_df, test_df = train_test_split(df, test_size=0.3, random_state=42)
    test_df, valid_df = train_test_split(test_df, test_size=0.5, random_state=42)
    return train_df, valid_df, test_df

# Dataset Class (unchanged)
class BrainTumorDataset(Dataset):
    def __init__(self, df, transform=None):
        self.df = df
        self.transform = transform
        self.img_transform = transforms.Compose([
            transforms.Normalize(mean=[0.485, 0.456, 0.406], 
                               std=[0.229, 0.224, 0.225])
        ])
        
    def __len__(self):
        return len(self.df)
    
    def __getitem__(self, idx, raw=False):
        row = self.df.iloc[idx]
        img = cv2.imread(row['image_filename'], cv2.IMREAD_COLOR)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, (Config.img_size, Config.img_size))
        
        mask = cv2.imread(row['mask_filename'], cv2.IMREAD_GRAYSCALE)
        mask = cv2.resize(mask, (Config.img_size, Config.img_size))
        mask = (mask > 0).astype(np.float32)
        
        if raw:
            return img, mask
        
        if self.transform:
            augmented = self.transform(image=img, mask=mask)
            img, mask = augmented['image'], augmented['mask']
        
        img = torch.from_numpy(img).permute(2, 0, 1).float() / 255.0
        img = self.img_transform(img)
        mask = torch.from_numpy(np.expand_dims(mask, axis=0)).float()
        
        return img, mask

# Augmentations (unchanged)
train_transform = A.Compose([
    A.ChannelDropout(p=0.3),
    A.RandomBrightnessContrast(p=0.3),
    A.ColorJitter(p=0.3),
])

# Diffusion Model (unchanged)
class DiffusionModel(nn.Module):
    def __init__(self, in_channels, num_timesteps=100):
        super().__init__()
        self.num_timesteps = num_timesteps
        self.betas = self.linear_beta_schedule(num_timesteps).to(Config.device)
        self.alphas = 1. - self.betas
        self.alpha_cumprod = torch.cumprod(self.alphas, dim=0)
        self.model = nn.Sequential(
            nn.Conv2d(in_channels, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.BatchNorm2d(64),
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.BatchNorm2d(64),
            nn.Conv2d(64, in_channels, kernel_size=3, padding=1)
        )
    
    def linear_beta_schedule(self, timesteps):
        scale = 1000 / timesteps
        beta_start = scale * 0.0001
        beta_end = scale * 0.02
        return torch.linspace(beta_start, beta_end, timesteps, dtype=torch.float32)
    
    def forward(self, x, t):
        sqrt_alpha_cumprod = torch.sqrt(self.alpha_cumprod[t])[:, None, None, None]
        sqrt_one_minus_alpha_cumprod = torch.sqrt(1.0 - self.alpha_cumprod[t])[:, None, None, None]
        noise = torch.randn_like(x)
        noisy_x = sqrt_alpha_cumprod * x + sqrt_one_minus_alpha_cumprod * noise
        predicted_noise = self.model(noisy_x)
        return predicted_noise, noise

# Full Model Architecture (unchanged)
class BrainTumorSegmentationModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.feature_extractor = mobilenet_v3_small(pretrained=True)
        self.feature_extractor.classifier = nn.Identity()
        self.feature_adapt = nn.Sequential(
            nn.Conv2d(576, 96, kernel_size=1),
            nn.ReLU(),
            nn.BatchNorm2d(96),
            nn.Upsample(size=(64, 64), mode='bilinear')
        )
        self.diffusion = DiffusionModel(96, Config.timesteps)
        self.swin = SwinTransformer(
            img_size=64,
            patch_size=4,
            in_chans=96,
            num_classes=1,
            embed_dim=96,
            depths=[2, 2, 6, 2],
            num_heads=[3, 6, 12, 24],
            window_size=8,
            drop_rate=0.1
        )
        self.channel_adjust = nn.Conv2d(768, 96, kernel_size=1)
        self.seg_head = nn.Sequential(
            nn.Conv2d(96, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.BatchNorm2d(64),
            nn.Upsample(size=(Config.img_size, Config.img_size), mode='bilinear'),
            nn.Conv2d(64, 1, kernel_size=1),
            nn.Sigmoid()
        )
    
    def forward_features(self, x):
        x = self.feature_extractor.features(x)
        x = self.feature_adapt(x)
        return x
    
    def forward(self, x, t=None):
        features = self.forward_features(x)
        if t is None:
            t = torch.randint(0, Config.timesteps, (x.size(0),), device=Config.device)
        predicted_noise, true_noise = self.diffusion(features, t)
        refined_features = features - predicted_noise
        swin_out = self.swin.forward_features(refined_features)
        swin_out = swin_out.permute(0, 3, 1, 2)
        swin_out = self.channel_adjust(swin_out)
        swin_out = nn.functional.interpolate(swin_out, size=refined_features.shape[2:], 
                                           mode='bilinear', align_corners=False)
        seg = self.seg_head(swin_out)
        return seg, predicted_noise, true_noise

# Training Utilities (unchanged)
class EarlyStopping:
    def __init__(self, patience=6, min_delta=0, weights_path='weights.pt'):
        self.patience = patience
        self.min_delta = min_delta
        self.counter = 0
        self.best_loss = float('inf')
        self.weights_path = weights_path

    def __call__(self, val_loss, model):
        if self.best_loss - val_loss > self.min_delta:
            self.best_loss = val_loss
            torch.save(model.state_dict(), self.weights_path)
            self.counter = 0
        else:
            self.counter += 1
            if self.counter >= self.patience:
                return True
        return False

    def load_weights(self, model):
        model.load_state_dict(torch.load(self.weights_path))

def dice_coeff(pred, target, e=1e-7):
    pred = (pred > 0.5).float()
    intersection = (pred * target).sum()
    return (2 * intersection + e) / (pred.sum() + target.sum() + e)

def BCE_dice(output, target, alpha=0.01):
    bce = nn.functional.binary_cross_entropy(output, target)
    soft_dice = 1 - dice_coeff(output, target)
    return bce + alpha * soft_dice

def training_loop(model, train_loader, valid_loader, epochs):
    optimizer = optim.Adam(model.parameters(), lr=Config.lr)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, patience=2, factor=0.2)
    early_stopping = EarlyStopping(patience=7)
    
    history = {'train_loss': [], 'val_loss': [], 'val_dice': []}
    
    for epoch in range(1, epochs + 1):
        model.train()
        train_loss = 0
        for img, mask in tqdm(train_loader):
            img, mask = img.to(Config.device), mask.to(Config.device)
            optimizer.zero_grad()
            seg_pred, _, _ = model(img)
            loss = BCE_dice(seg_pred.squeeze(1), mask.squeeze(1))
            loss.backward()
            optimizer.step()
            train_loss += loss.item()
        
        model.eval()
        val_loss = 0
        val_dice = 0
        with torch.no_grad():
            for img, mask in valid_loader:
                img, mask = img.to(Config.device), mask.to(Config.device)
                seg_pred, _, _ = model(img)
                loss = BCE_dice(seg_pred.squeeze(1), mask.squeeze(1))
                val_loss += loss.item()
                val_dice += dice_coeff(seg_pred.squeeze(1), mask.squeeze(1)).item()
        
        train_loss /= len(train_loader)
        val_loss /= len(valid_loader)
        val_dice /= len(valid_loader)
        
        history['train_loss'].append(train_loss)
        history['val_loss'].append(val_loss)
        history['val_dice'].append(val_dice)
        
        print(f'Epoch {epoch}/{epochs} | Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f} | Val Dice: {val_dice:.4f}')
        
        scheduler.step(val_loss)
        if early_stopping(val_loss, model):
            print("Early stopping triggered")
            early_stopping.load_weights(model)
            break
    
    return history

# New Evaluation Metrics Functions
def plot_confusion_matrix(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=['No Tumor', 'Tumor'], 
                yticklabels=['No Tumor', 'Tumor'])
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.show()

def plot_roc_curve(y_true, y_scores):
    fpr, tpr, thresholds = roc_curve(y_true, y_scores)
    roc_auc = auc(fpr, tpr)
    
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc="lower right")
    plt.show()

def plot_precision_recall_curve(y_true, y_scores):
    precision, recall, thresholds = precision_recall_curve(y_true, y_scores)
    average_precision = average_precision_score(y_true, y_scores)
    
    plt.figure(figsize=(8, 6))
    plt.plot(recall, precision, color='blue', lw=2, 
             label=f'Precision-Recall curve (AP = {average_precision:.2f})')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Precision-Recall Curve')
    plt.legend(loc="lower left")
    plt.show()

def generate_heatmap(model, image_tensor):
    model.eval()
    with torch.no_grad():
        seg_pred, _, _ = model(image_tensor.unsqueeze(0).to(Config.device))
        heatmap = seg_pred.squeeze().cpu().numpy()
    
    plt.figure(figsize=(8, 6))
    plt.imshow(heatmap, cmap='hot', interpolation='nearest')
    plt.colorbar()
    plt.title('Model Confidence Heatmap')
    plt.axis('off')
    plt.show()

def evaluate_model(model, test_loader):
    model.eval()
    all_preds = []
    all_targets = []
    all_scores = []
    
    with torch.no_grad():
        for img, mask in tqdm(test_loader):
            img, mask = img.to(Config.device), mask.to(Config.device)
            seg_pred, _, _ = model(img)
            
            # Flatten predictions and targets
            pred_flat = (seg_pred > 0.5).float().view(-1).cpu().numpy()
            target_flat = mask.view(-1).cpu().numpy()
            scores_flat = seg_pred.view(-1).cpu().numpy()
            
            all_preds.extend(pred_flat)
            all_targets.extend(target_flat)
            all_scores.extend(scores_flat)
    
    # Convert to numpy arrays
    all_preds = np.array(all_preds)
    all_targets = np.array(all_targets)
    all_scores = np.array(all_scores)
    
    # Plot metrics
    plot_confusion_matrix(all_targets, all_preds)
    plot_roc_curve(all_targets, all_scores)
    plot_precision_recall_curve(all_targets, all_scores)
    
    # Print classification report
    print(classification_report(all_targets, all_preds, target_names=['No Tumor', 'Tumor']))
    
    return all_preds, all_targets, all_scores

# Visualization (updated with heatmap)
def visualize_results(history):
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history['train_loss'], label='Train Loss')
    plt.plot(history['val_loss'], label='Val Loss')
    plt.legend()
    plt.title('Training and Validation Loss')
    
    plt.subplot(1, 2, 2)
    plt.plot(history['val_dice'], label='Val Dice')
    plt.legend()
    plt.title('Validation Dice Coefficient')
    plt.show()

def visualize_predictions(model, loader, n_examples=4):
    model.eval()
    fig, axs = plt.subplots(n_examples, 3, figsize=(15, n_examples*5))
    
    with torch.no_grad():
        for i in range(n_examples):
            while True:
                idx = np.random.randint(len(loader.dataset))
                img, mask = loader.dataset.__getitem__(idx, raw=True)
                if mask.sum() > 0:  # Only show examples with tumors
                    break
            
            img_tensor, _ = loader.dataset[idx]
            img_tensor = img_tensor.unsqueeze(0).to(Config.device)
            pred, _, _ = model(img_tensor)
            pred = (pred.squeeze().cpu().numpy() > 0.5).astype(np.float32)
            
            axs[i,0].imshow(img)
            axs[i,0].set_title('Original Image')
            axs[i,0].axis('off')
            
            axs[i,1].imshow(img)
            axs[i,1].imshow(mask, alpha=0.3, cmap='Reds')
            axs[i,1].set_title('Ground Truth')
            axs[i,1].axis('off')
            
            axs[i,2].imshow(img)
            axs[i,2].imshow(pred, alpha=0.3, cmap='Greens')
            axs[i,2].set_title('Prediction')
            axs[i,2].axis('off')
    
    plt.tight_layout()
    plt.show()
    
    # Generate heatmap for the last example
    generate_heatmap(model, img_tensor.squeeze(0).cpu())

# Main Execution (updated)
def main():
    # Load data
    train_df, valid_df, test_df = load_data()
    
    # Create datasets and loaders
    train_dataset = BrainTumorDataset(train_df, train_transform)
    valid_dataset = BrainTumorDataset(valid_df)
    test_dataset = BrainTumorDataset(test_df)
    
    train_loader = DataLoader(train_dataset, batch_size=Config.batch_size, shuffle=True)
    valid_loader = DataLoader(valid_dataset, batch_size=Config.batch_size)
    test_loader = DataLoader(test_dataset, batch_size=1)
    
    # Initialize model
    model = BrainTumorSegmentationModel().to(Config.device)
    
    # Train model
    history = training_loop(model, train_loader, valid_loader, Config.num_epochs)
    
    # Visualize results
    visualize_results(history)
    visualize_predictions(model, test_loader)
    
    # Evaluate model with metrics
    print("\nEvaluating model on test set...")
    evaluate_model(model, test_loader)
    
    # Save model
    torch.save(model.state_dict(), Config.save_path)
    print(f"Model saved to {Config.save_path}")

if __name__ == "__main__":
    main()
```
